```python
##Doing PCA to get Results
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA 
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
pca = PCA(2) 
pca_data = pd.DataFrame(pca.fit_transform(data_final.drop(['clusters'],axis=1)),columns=['PC1','PC2'])
pca_data['clusters']=predictions
```

```python
#RESULTS Analysis
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
sns.scatterplot(data=pca_data,x='PC1',y='PC2',hue='clusters',palette='Set2' , alpha = 0.9)
plt.title('Music Recommendation after PCA')
```

```python
def optimise_k_means(data max_k):
    means=[]
    inertias=[]
    for k in range(1, max_k):
        kmeans=KMeans(n_clusters=k)
        kmeans.fit(data_final)
        
        means.append(k)
        inertias.append(kmeans.inertia_)
    #generate the elbow plot
    fig=plt.subplots(figsize=(105))
    plt.plot(means,inertias, 'o-')
    plt.xlabel('number of clusters')
    plt.ylabel('Inertia')
    plt.grid(True)
    plt.show()
```
